# PCA

##  Normalize the Data

```python
test_pca =churn_df[['timely_response', 'timely_fixes', 'timely_replacements','reliability','options', 'respectful_response','courteous_exchange', 'evidence_of_active_listening']]
test_pca_normalized=(test_pca-test_pca.mean())/test_pca.std()
```

```python
pca=PCA(n_components=test_pca.shape[1])
```

```python
pca.fit(test_pca_normalized)
```

```python
test_pca=pd.DataFrame(pca.transform(test_pca_normalized),
columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5','PC6','PC7','PC8'])                      
```

```python
pca.fit(test_pca_normalized)
```

```python
test_pca2=pd.DataFrame(pca.transform(test_pca_normalized),columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5','PC6','PC7','PC8'])
```

```python
loadings =pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5','PC6','PC7','PC8'], index=test_pca_normalized.columns)
loadings
```

```python
cov_matrix=np.dot(test_pca_normalized.T, test_pca_normalized)/test_pca.shape[0]
eigenvalues=[np.dot(eigenvector.T,np.dot(cov_matrix,eigenvector))for eigenvector in pca.components_]
```

```python
plt.plot(eigenvalues)
plt.xlabel('number of components')
plt.ylabel('eigenvalues')
plt.axhline(y=1, color="red")
plt.show()
```
